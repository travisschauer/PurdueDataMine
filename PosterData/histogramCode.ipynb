```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#create dataframe
nodeDF = pd.read_excel('nodes.xlsx')
```

```python
len(nodeDF[nodeDF['eigencentrality'] < .1])
```

```python
cleanedDF = nodeDF[nodeDF['eigencentrality'] >= .1]
```

```python
# Create centrality histogram
counts, bins, patches = plt.hist(cleanedDF['eigencentrality'], bins=10, color='#FEC938')
plt.xlim(0, 1)
plt.ylim(0, max(counts) + 2)
plt.title("Eigenvector Centrality of the Instagram Network")
plt.xlabel("Centrality Score")
plt.ylabel("# of Nodes")

# Add labels to the top of each non-zero column
for i in range(len(counts)):
    if counts[i] > 0:
        plt.text(bins[i] + 0.05, counts[i] + .5, str(int(counts[i])), ha='center')

# Adjust axis labels
plt.yticks([])
plt.xticks([i/10 for i in range(11)])

plt.show()
```
